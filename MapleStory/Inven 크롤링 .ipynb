## 웹 서버 요청

```python
def web_server_request(code_number, page_number):
    # 메이플 인벤 직업 게시판 사이트
    url = "https://www.inven.co.kr/board/maple/{}?p={}".format(code_number,page_number)

    resp = requests.get(url)
    return resp
```

### 전사

```python
warrior_text = []

for page_number in range(15, 154):
    warrior = web_server_request(2294, page_number)
    warrior_text.append(warrior.text)
```

## HTML 파싱

```python
soup_list = []

# BeautifulSoup 함수로 HTML 문서 구조를 파싱
for text in warrior_text:
    soup = BeautifulSoup(text, 'html.parser')
    soup_list.append(soup)
    # 파싱한 내용을 담고 있는 객체의 자료형 확인
    # print(type(soup))
```

## 게시물 목록

```python
post_num_list = []
post_category_list = []
post_title_list = []
post_writer_list = []
post_date_list = []
job_list = []
view_list = []
reco_list = []
comment_list = []

for soup in soup_list:
    post_list_tag = soup.find("tbody")
    post_list_tags = post_list_tag.find_all('tr')
    
    job = soup.find("span", attrs = {'class' : 'main'}).get_text()[9:13].strip()

    for tr_tag in post_list_tags:
        job_list.append(job)

        post_num = tr_tag.find("td", attrs = {'class': 'num'}).find("span").get_text()
        post_num_list.append(post_num)

        post_category = tr_tag.find("span", attrs = {'class' : 'category'}).get_text().strip("[""]")
        post_category_list.append(post_category)

        post_title = tr_tag.find("a", attrs = {'class' : 'subject-link'}).get_text()[10:].strip()   # category 및 띄어쓰기 없애주기
        post_title_list.append(post_title)

        post_writer = tr_tag.find("span", attrs = {'class' : 'layerNickName'}).get_text()
        post_writer_list.append(post_writer)

        post_date = tr_tag.find("td", attrs = {'class' : 'date'}).get_text()
        post_date_list.append(post_date)

        view = tr_tag.find("td", attrs = {'class' : 'view'}).get_text()
        view_list.append(view)

        reco = tr_tag.find("td", attrs = {'class' : 'reco'}).get_text()
        reco_list.append(reco)

        comment = tr_tag.find("span", attrs = {'class' : 'con-comment'}).get_text().strip("[""]")
        comment_list.append(comment)


print(post_num_list)
print(post_category_list)
print(post_title_list)
print(post_writer_list)
print(post_date_list)
print(job_list)
print(view_list)
print(reco_list)
print(comment_list)
```

## 데이터 수집

```python
dict_data = {
    '번호' : post_num_list,
    '직업' : job_list,
    '분야' : post_category_list,
    '제목' : post_title_list,
    '작성자' : post_writer_list,
    '날짜' : post_date_list,
    '조회수' : view_list,
    '추천수' : reco_list,
    '댓글수' : comment_list
}

df_data = pd.DataFrame(dict_data)
df_data
```
