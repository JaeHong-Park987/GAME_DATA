## 웹 서버 요청

```python
def web_server_request(code_number):
    # 메이플 인벤 직업 게시판 사이트
    url = "https://www.inven.co.kr/board/maple/{}".format(code_number)

    resp = requests.get(url)
    return resp
```

```python
warrior = web_server_request(2294)
wizard = web_server_request(2295)
archer = web_server_request(2296)
thief = web_server_request(2297)
pirate = web_server_request(2298)
```

## HTML 파싱

```python
# 5개의 직업 서버 응답 텍스트 리스트 생성
job_text = [warrior.text, wizard.text, archer.text, thief.text, pirate.text]

# 파싱한 내용을 담을 리스트 생성
soup_list = []

# BeautifulSoup 함수로 HTML 문서 구조를 파싱
for text in job_text:
    soup = BeautifulSoup(text, 'html.parser')
    soup_list.append(soup)
    # 파싱한 내용을 담고 있는 객체의 자료형 확인
    print(type(soup))
```

## 게시판 제목

```python
for soup in soup_list:
    # <span class="main">메이플스토리 인벤 전사 게시판</span>
    # span 태그의 class 속성값을 활용하여 게시판 제목이 포함되어 있는 요소를 찾습니다.
    board_title_tag = soup.find("span", attrs = {'class' : 'main'})
    board_title = board_title_tag.get_text()
    print("게시판 제목 : ",board_title)
```
